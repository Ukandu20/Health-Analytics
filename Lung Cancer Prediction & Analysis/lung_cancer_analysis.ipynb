## üõ† Data Cleaning and Preparation  

### ‚ùì How do we ensure data quality before analysis?  
- Are there **missing values** in the dataset, and how do we handle them?  
- Why do we need to **categorize age into groups** instead of using continuous values?  
- Why is it important to **convert categorical variables (e.g., Pulmonary Disease) into numeric format** for analysis?  

### üîç Data Cleaning Steps  
1. **Handling Missing Values:**  
   - **Dropped all rows** with missing values to ensure a clean dataset.  
   - Alternative approaches (e.g., **imputation**) could be used, but dropping is simpler if missing data is minimal.  

2. **Categorizing Age into Groups:**  
   - Binned age into **10-year intervals** (`0-10`, `11-20`, ..., `91-100`).  
   - Helps in **trend analysis** across different age ranges instead of using raw age values.  

3. **Converting Pulmonary Disease to Numeric:**  
   - Transformed **'Yes' ‚Üí 1** and **'No' ‚Üí 0** for easier numerical analysis.  
   - This step is **crucial for machine learning models** and statistical tests.  

4. **Creating a Numeric-Only DataFrame:**  
   - Retained only **numerical columns**, making it easier for **correlation analysis and modeling**.  

### üìå Observations & Next Steps  
- How does the **distribution of missing values** affect the dataset? Would **imputation** be a better approach?  
- Does categorizing age improve **interpretability**, or would keeping it continuous be more useful?  
- Is there any **data imbalance** in the Pulmonary Disease column after conversion?  
- Now that the dataset is cleaned, we can move on to **exploratory data analysis (EDA)** and **statistical testing**.  

This data cleaning step ensures that the dataset is **well-structured, ready for analysis, and compatible with numerical methods.**  


```python

# Remove rows with missing values
df = df.dropna()

# Define age brackets dynamically
age_brackets = list(range(0, 101, 10))  # 0-10, 11-20, ..., 91-100
age_labels = [f"{age_brackets[i]}-{age_brackets[i+1]-1}" for i in range(len(age_brackets)-1)]

# Create age groups
df['AGE_GROUP'] = pd.cut(df['AGE'], bins=age_brackets, labels=age_labels, include_lowest=True)

# Convert 'PULMONARY_DISEASE' column to integer (No ‚Üí 0, Yes ‚Üí 1)
if 'PULMONARY_DISEASE' in df.columns:
    df['PULMONARY_DISEASE'] = df['PULMONARY_DISEASE'].map({'No': 0, 'Yes': 1, 'YES': 1, 'NO': 0}).fillna(0).astype(int)

# Create a numeric-only DataFrame (drop non-numeric columns)
df_numeric = df.select_dtypes(include=[np.number])

df_numeric.head()
```

```python
# Define plot settings
plot_settings = {
    'figsize': (30, 15),  # Set figure size
    'barplot_title': 'Correlation of Features with Target Variable',
    'barplot_xlabel': 'Features',
    'barplot_ylabel': 'Correlation Coefficient',
    'heatmap_title': 'Heatmap of Correlations',
    'xticks_rotation': 45
}

# Function to plot the correlation of all features with a target column using subplots(Bar chart and Heatmap)
def plot_correlations(df_numeric, target_column):
    """
    Plot the correlation of all features with a target column using a bar chart and heatmap.

    Args:
        df_numeric (pd.DataFrame): The DataFrame containing numerical data.
        target_column (str): The name of the target column.
    """

    # Compute correlations
    correlations = df_numeric.corrwith(df_numeric[target_column]).sort_values(ascending=False)
    corr_matrix = df_numeric.corr()

    # Create subplots
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=plot_settings['figsize'])

    # Correlation bar plot
    sns.barplot(
    x=correlations.index,
    y=correlations.values,
    ax=axes[0],
    hue=correlations.index, 
    palette="coolwarm",
    legend=False  
    )

    axes[0].set_title(plot_settings['barplot_title'])
    axes[0].set_xlabel(plot_settings['barplot_xlabel'])
    axes[0].set_ylabel(plot_settings['barplot_ylabel'])
    axes[0].tick_params(axis='x', rotation=plot_settings['xticks_rotation'])

    # Heatmap of correlations
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True, fmt=".2f", ax=axes[1])
    axes[1].set_title(plot_settings['heatmap_title'])

    # Improve layout
    plt.tight_layout()
    plt.show()

    # Print correlation values
    print(f"\nTop Correlations with {target_column}:")
    print(correlations.head())

# Usage
plot_correlations(df_numeric, 'PULMONARY_DISEASE')
```

```python
import scipy.stats as stats

# ‚úÖ Step 1: Create a contingency table
contingency_table = pd.crosstab(df['SMOKING'], df['PULMONARY_DISEASE'])

# ‚úÖ Step 2: Perform Chi-Square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# ‚úÖ Step 3: Print results
print("üìä Chi-Square Test Results:")
print(f"üîπ Chi-Square Statistic: {chi2:.2f}")
print(f"üîπ Degrees of Freedom: {dof}")
print(f"üîπ P-Value: {p:.5f}")

# ‚úÖ Step 4: Interpretation
alpha = 0.05
if p < alpha:
    print("üö® **Statistically Significant!** There is a strong association between smoking and pulmonary disease.")
else:
    print("‚úÖ **Not Significant.** No strong evidence of an association between smoking and pulmonary disease.")
```
